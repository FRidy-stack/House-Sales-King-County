```python
X = data[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement',
       'sqft_living15', 'sqft_lot15']]
```

```python
X
```

```python
Y = data['price']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.25, random_state = 101);
# random_state is seed value
```

<h1> Normalization, scaling of data, (Standdarization)

```python
# so that data reach with one another, 1000000 -> 1 M, 1000-> 1 K, ranging in data
```

```python
# so mean become 0 and std = 1
# methods of normalization, z value , etc
# bell shape curve
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
```

```python
# pip install sklearn
```

```python
X_train_std = std.fit_transform(X_train)
```

```python
X_train
```

```python
X_train_std
```

```python
X_train_std.shape
```

```python
X_test_std = std.transform(X_test)
```

```python
X_test_std
```

```python
Y_train
```

```python
Y_test
```

<h1> Model Training

```python
# Linear regression
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

<h1> Don't use standard scaler (normalized version) use normal

```python
lr.fit(X_train_std, Y_train)
```

```python
Y_pred = lr.predict(X_test_std)
```

```python
Y_test
```

```python
Y_pred
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mean_absolute_error(Y_test, Y_pred)
```
